```python
#imported the libraries
import pandas as pd
import datetime 
```

```python
#loaded the dataset 
data_df=pd.read_csv('sales_data.csv')
```

```python
data_df.head()
```

```python
#checking the null values
data_df.isnull().sum()
```

```python
#cleaning the column 
data_df.columns=data_df.columns.str.capitalize().str.replace(' ','_')
```

```python
data_df.columns
```

```python
#drop the null row
data_df.dropna(subset=['Order_id','Product','Quantity_ordered','Price_each','Order_date','Purchase_address'],how='all',inplace=True)
```

```python
data_df
```

```python
#fillling the null values using the mean 
data_df.fillna(data_df['Price_each'].mean(),inplace=True)
```

```python
#replaceing the category data to the int in quantity ordered column
data_df.iloc[2]=data_df.iloc[2].replace('two',2)
```

```python
data_df
```

```python
#datetime formatting 
data_df.iloc[:,4]=pd.to_datetime(data_df.iloc[:,4],errors='coerce')
```

```python
data_df
```

```python
#created the street,city,state columns
data_df[['Street','City','State']]=data_df.iloc[:,5].str.split(',',expand=True)
```

```python
data_df
```

```python
#convert to numeric data
data_df.iloc[:,3:4]=data_df.iloc[:,3:4].astype(float)
```

```python
data_df['Quantity_ordered']=pd.to_numeric(data_df['Quantity_ordered'],errors='coerce')
```

```python
data_df.dtypes
```

```python
#cleaned the data
data_df['Total_price']=(data_df.iloc[:,2]*data_df['Price_each']).round(2)
```

```python
data_df
```

```python
#convert the df to the csv file
result=data_df.to_csv('reault_data.csv')
```
